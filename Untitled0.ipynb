```
from ipywidgets import interact
import numpy as np

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()

t = np.linspace(0, 20, 100)
y = 3*np.exp(-0.2*t) - 3./4.*np.exp(-t)

p = figure(title="A differenciálegyenlet megoldása", plot_height=300, plot_width=600, y_range=(-5,5),
           background_fill_color='#efefef')
r = p.line(t, y, color='blue', line_width=2, alpha=1)
x_axis = p.line(t, 0, color='black', line_width=1, alpha=1)

def update(C=3):
    r.data_source.data['y'] = C*np.exp(-0.2*t) - 3./4.*np.exp(-t)
    push_notebook()
    
show(p, notebook_handle=True)
interact(update, C=(-3, 5, 0.1));
```
